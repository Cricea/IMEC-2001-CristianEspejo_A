### Ejercicio 2: Gráficas

En la carpeta `data` dentro de `week2` se dispone el archivo `solar_production.csv` que contiene datos de producción de la planta solar fotovoltaica ubicada en el edificio Santo Domingo de la Universidad de los Andes.

<div class="alert alert-block alert-warning">
    
Utilizando la librería `pandas` y `matplotlib.pyplot`, escriba un algoritmo que permita realizar tres gráficas diferentes de su interés.

</div>

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
import matplotlib.cm

route = './data/solar_production.csv'
plt.ion()
file = pd.read_csv(filepath_or_buffer=route,
                   delimiter=';')
años=pd.DatetimeIndex(file["Datetime"]).year
file["year"]=años
plt.figure()
file.plot.scatter(x="Tmod (C)", y='POA (W/m2)',color='darkorange')
plt.title('Tmod vs POA')
plt.xlabel('Tmod (C)')
plt.ylabel('POA (W/m2)')
plt.show()

ax = plt.axes(projection="3d")
x_data= file["Airmass Absolute (ad)"]
z_data=file["POA (W/m2)"]
y_data=file["Extraterrestrial Radiation (W/m2)"]
ax.scatter(x_data,y_data,z_data,cmap="plasma")
plt.title('Airmass Absolute vs Extraterrestrial Radiation vs POA ')
ax.set_xlabel('Airmass Absolute (ad)')
ax.set_ylabel('Extraterrestrial Radiation (W/m2)')
ax.set_zlabel('POA (W/m2)')
plt.show()

# grafico 3d
ax = plt.axes(projection="3d")
x_data= file["Azimuth (deg)"]
y_data=file["Elevation (deg)"]
z_data=file["POA (W/m2)"]
ax.scatter(x_data,y_data,z_data,cmap="hot")
plt.title('Azimuth vs. elevation vs POA ')
ax.set_xlabel('Azimuth (deg)')
ax.set_ylabel('Elevation (deg)')
ax.set_zlabel('POA (W/m2)')
plt.set_cmap("hot")
plt.show()
```
